# One Hot Encoding with sciket Learn Library

```python
df
```

```python
df.place.value_counts()
```

```python
df.place.value_counts().shape
```

```python
df.place.unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False).set_output(transform='pandas')
```

```python
ohetransform = ohe.fit_transform(df[['place']])
```

```python
ohetransform
```

```python
oheencoded_df = pd.concat([df,ohetransform], axis=1)
```

```python
oheencoded_df
```
