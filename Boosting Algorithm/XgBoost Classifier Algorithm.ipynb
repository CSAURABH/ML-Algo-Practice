```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import xgboost
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv(r'C:\Users\MY MAC\OneDrive\Desktop\Machine Learning Algo Practice\Datasets\Churn_Modelling.csv')
```

```python
df.drop(columns=['RowNumber', 'CustomerId', 'Surname'], inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns= ['Geography'], drop_first=True, dtype=int)
```

```python
df
```

```python
df = pd.get_dummies(df, columns= ['Gender'], drop_first=True, dtype=int)
```

```python
df.head()
```

```python
X = df.drop('Exited', axis = 1)
y = df['Exited']
```

```python
X.rename(columns={'Geography_Germany' : 'Germany', 'Geography_Spain' : 'Spain', 'Gender_Male' : 'Male'}, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.03)
```

```python
## Hyper Parameter Optimization

params={
 "learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
 "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
 "min_child_weight" : [ 1, 3, 5, 7 ],
 "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ]
    
}
```

```python
xgclassifier = xgboost.XGBClassifier()
```

```python
random_search = RandomizedSearchCV(xgclassifier, param_distributions=params, n_iter=5, scoring='roc_auc', n_jobs=-1, cv=5, verbose=3)
```

```python
random_search.fit(X_train, y_train)
```

```python
y_pred = random_search.predict(X_test)
```

```python
y_pred
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
random_search.best_params_
```

```python
random_search.best_estimator_
```
